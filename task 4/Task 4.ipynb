```python
#Write a python program to scrape mentioned details from dineout.co.in : i) Restaurant name
# ii) Cuisine iii) Location iv) Ratings v) Image URL.

page3=requests.get("https://www.dineout.co.in/chennai-restaurants")
page3
```

```python
soup2=BeautifulSoup(page3.content)
soup2
```

```python
restaurantnames = []

        
for i in soup2.find_all('div', class_='restnt-info'):
    restaurantnames.append(i.text)

    
restaurantnames
```

```python
cuisines = []

for i in soup2.find_all('a', attrs={'data-w-onclick': 'stopClickPropagation|w1-restarant'}):
    cuisines.append(i.text)

cuisines
```

```python
locations = []

for i in soup2.find_all('div', class_='restnt-loc ellipsis'):
    locations.append(i.text)
locations
```

```python
ratings = []

for i in soup2.find_all('div', class_='restnt-rating'):
    ratings.append(i.text)
    
ratings
```

```python
ImageUrl = []

for i in soup2.find_all('img', class_='no-img'):
    ImageUrl.append(i['data-src'])
    
ImageUrl
```

```python
#Write s python program to display list of respected former finance minister of India(i.e.
# Name , Term of office) from https://presidentofindia.nic.in/former-presidents.htm and make data frame.

page4=requests.get("https://presidentofindia.nic.in/former-presidents")
page4
```

```python
soup3=BeautifulSoup(page4.content)
soup3
```

```python
PresidentNames=[]

for i in soup3.find_all('div', class_='desc-sec'):
    PresidentNames.append(i.h3.text)
    
PresidentNames
```

```python
TermOffice=[]

for i in soup3.find_all('div', class_='desc-sec'):
    TermOffice.append(i.h5.text)
    
TermOffice
```

```python
table=pd.DataFrame({'PresidentNames':PresidentNames,'TermOffice':TermOffice})
table
```
