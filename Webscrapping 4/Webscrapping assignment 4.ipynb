```python
import selenium
import pandas as pd
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.common.keys import Keys

import time
```

```python
# Q1
driver=webdriver.Chrome()
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
detailtag = driver.find_elements(By.XPATH, '//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td')
songs = []

for row in detailtag:
    songs.append(row.text)

songs
```

```python
Rank = []
Name = []
Artist = []
Views = []
UploadDate = []

for i in range(0, len(songs), 5):
    rank = i // 5 + 1
    Name.append(songs[i])
    Artist.append(songs[i + 1])
    Views.append(songs[i + 2])
    UploadDate.append(songs[i + 3])
    Rank.append(rank)
```

```python
df=pd.DataFrame({
    "Rank":Rank,
    "Name":Name,
    "Artist":Artist,
    "Upload Date":UploadDate,
    "Views":Views
})

df
```

```python
driver.quit()
```

```python
# Q5 
driver=webdriver.Chrome()
driver.get("https://www.billboard.com")
```

```python
hot100tag=driver.find_element(By.XPATH,'//a[@class="c-link  lrv-a-unstyle-link lrv-u-color-brand-primary:hover lrv-a-hover-effect lrv-u-whitespace-nowrap lrv-u-color-grey-dark"]')
hot100tag.click()
```

```python
Nametags=driver.find_elements(By.XPATH,'//ul[@class="lrv-a-unstyle-list lrv-u-flex lrv-u-height-100p lrv-u-flex-direction-column@mobile-max"]/li/h3')
infotags=driver.find_elements(By.XPATH,'//ul[@class="lrv-a-unstyle-list lrv-u-flex lrv-u-height-100p lrv-u-flex-direction-column@mobile-max"]/li/span')

data=[]
SongName=[]
for i in Nametags:
    SongName.append(i.text)
for i in infotags:
        data.append(i.text)
        
```

```python
artist = []
lastweek = []
peakposition = []
weeksonchart = []

for index, value in enumerate(data):
    if index % 4 == 0:
        artist.append(value)
    elif index % 4 == 1:
        lastweek.append(value)
    elif index % 4 == 2:
        peakposition.append(value)
    else:
        weeksonchart.append(value)
```

```python
df5=pd.DataFrame({
    "Song Name":SongName,
    "Artist":artist,
    "Last week":lastweek,
    "Peak Position":peakposition,
    "Weeks on Chart":weeksonchart
})

df5
```

```python
# Q8
driver=webdriver.Chrome()
driver.get("https://archive.ics.uci.edu")
```

```python
dataset=driver.find_element(By.XPATH,'//a[@class="btn-primary btn"]')
dataset.click()
```

```python
nametag=driver.find_elements(By.XPATH,'//a[@class="link-hover link text-xl font-semibold"]')
infotag=driver.find_elements(By.XPATH,'//span[@class="truncate"]')
DatasetName=[]
datainfo=[]
for i in nametag:
    DatasetName.append(i.text)
for i in infotag:
    datainfo.append(i.text)
    
datainfo
```

```python
Datatype = []
Task = []
Instances = []
Attribute = []

counter = 0 

for index, value in enumerate(datainfo):
    if counter >= 40:
        break
    if index % 4 == 0:
        Attribute.append(value)
    elif index % 4 == 1:
        Datatype.append(value)
    elif index % 4 == 2:
        Task.append(value)
    else:
        Instances.append(value)
    counter += 1  
```

```python
df8=pd.DataFrame({
    "Dataset Name":DatasetName,
    "Datatype":Datatype,
    "Task":Task,
    "No of Instances":Instances,
    "No of Attribute":Attribute
})

df8
```
