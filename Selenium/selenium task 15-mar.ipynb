```python
#Q6: Write a python program to scrape data for Top 1000 Quotes of All Time.

driver6=webdriver.Chrome()
driver6.get("https://www.azquotes.com")
```

```python
topquotes=driver6.find_element(By.XPATH,'/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a')
topquotes.click()
```

```python
quote=[]
author=[]
typeofquote=[]
```

```python
pages = driver6.find_elements(By.XPATH, '//div[@class="pager"]/li')

for i in pages:
    quotetags = driver6.find_elements(By.XPATH, '//a[@class="title"]')
    authortags= driver6.find_elements(By.XPATH, '//div[@class="author"]/a')
    typetags= driver6.find_elements(By.XPATH, '//div[@class="tags"]')
    for j in quotetags:
        title = j.text
        quote.append(title)
    for k in authortags:
        auth=k.text
        author.append(auth)
    for l in typetags:
        typ=l.text
        typeofquote.append(typ)
```

```python
df6 = pd.DataFrame({
    'Quote': quote,
    'Author': author,
    'type': typeofquote})


DF6_1000 = df6.head(1000)

DF6_1000
```
